### Imports

```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets
import matplotlib.colors as colors
from itertools import cycle

from sklearn.metrics import silhouette_score

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
```

```python
def dunn_index(X,labels):
    K = len(set(labels) - set([-1]))
    if K < 2:
        return -1
    d = len(X[0])

    ICD = np.zeros(K)
    centers = np.zeros([K,d])

    for i in range(K):
        Xk = X[labels == i]
        center = np.mean(Xk,axis=0)
        centers[i:i+1] = center
        inter_cluster_dist = 2*np.mean(np.sum((Xk-center)**2,axis=1)**0.5)
        ICD[i] = inter_cluster_dist

    Intrs_CD = np.sqrt(-2 * np.dot(centers, centers.T) + np.sum(centers**2,axis=1) + np.sum(centers**2, axis=1)[:, np.newaxis])
    res = min([Intrs_CD[i,j] for i in range(1,K) for j in range(i)]) / max(ICD)
    return res
```

```python
def davies_bouldian_index(X,labels):
    K = len(set(labels) - set([-1]))
    if K < 2:
        return -1

    d = len(X[0])

    ICD = np.zeros(K)
    centers = np.zeros([K,d])

    for i in range(K):
        Xk = X[labels == i]
        center = np.mean(Xk,axis=0)
        centers[i:i+1] = center
        inter_cluster_dist = 2*np.mean(np.sum((Xk-center)**2,axis=1)**0.5)
        ICD[i] = inter_cluster_dist

    Intrs_CD = np.sqrt(-2 * np.dot(centers, centers.T) + np.sum(centers**2,axis=1) + np.sum(centers**2, axis=1)[:, np.newaxis])

    res = np.zeros(K)
    for i in range(K):
        m = 0
        for j in range(K):
            if i!=j:
                m = max(m,ICD[j]/Intrs_CD[i,j])
        res[i] = m + ICD[i]
    
    return np.mean(res)
```

```python
def plotClustering(X, plt_labels,show_noice = False):
    plt_colors = cycle(colors.cnames.keys())
    _ = next(plt_colors)
    plt_K = np.unique(plt_labels).size

    for k in range(plt_K):
        color = next(plt_colors)
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')

    if show_noice:
        mask = (plt_labels == -1)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor="Black", marker='o')

    plt.show()
    print(f"Silhouette Coefficient: {silhouette_score(X,plt_labels)}")
    print(f"Dunn index: {dunn_index(X,plt_labels)}")
    print(f"Davies Bouldian index {davies_bouldian_index(X,plt_labels)}")
```

## Parametr n_cluster
c) Jakie znaczenie ma parametr n_clusters? Czym różni się n_clusters =None od n_clusters = 3?

n_clusters : int, instance of sklearn.cluster model, default 3
Number of clusters after the final clustering step, which treats the subclusters from the leaves as new samples.

1. None : the final clustering step is not performed and the subclusters are returned as they are.
2. sklearn.cluster Estimator : If a model is provided, the model is fit treating the subclusters as new samples and the initial data is mapped to the label of the closest subcluster.
3. int : the model fit is AgglomerativeClustering with n_clusters set to be equal to the int.

## Normalizacja

```python
def normalize(X):
    min_ = np.min(X,0)
    max_ = np.max(X,0)
    return (X - min_) / (max_-min_)
```
